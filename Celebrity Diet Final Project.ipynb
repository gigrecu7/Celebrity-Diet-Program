```python
import sys

def main():
    #here the main function is defined and this function calls all the other functions
    #print("main")
    name, maleOrFemale, actorChoice, age, height, weight = userInput()
    name, dietAnswer, caloricIntake, recFood, dietLength, healthRisk, dietFeedback = dietProcessing(name, maleOrFemale, actorChoice, age, height, weight)
    dietOutput(name, dietAnswer, caloricIntake, recFood, dietLength, healthRisk, dietFeedback)


def userInput():
    #print("input")
    print("Hello and Welcome to Celebrity Fitness and Nutrition")
    input("Press ENTER to continue")
    name = input ("What is your name? ")
    print("Hello " + name + " would you like to transform your body to look like the celebrity you want" + "?")
    invalidInput = True
    #here invalidInput if true, then user stays in the while loop
    while invalidInput:
        yOrN = input("enter 'yes' or 'no' ")
        yOrN = yOrN.lower()
        if yOrN == "yes":
            print("Great, let's get started then")
            invalidInput = False
        # here invalidInput if false, it breaks while loop and moves to next question
        elif yOrN == "no":
            print("That's okay, come back soon!")
            sys.exit()
        else:
            print("Whoops!  Incorrect input.  Please enter, 'yes' or 'no'")
    input("Press ENTER to continue")
    invalidInput = True
    while invalidInput:
        maleOrFemale = input("Are you a male or female? ")
        maleOrFemale = maleOrFemale.lower()
        if maleOrFemale != "male" and maleOrFemale != "female":
            print("Whoops!  Incorrect input.  Please enter, 'male' or 'female'")
        else:
            invalidInput = False

    invalidInput = True
    while invalidInput:
        if maleOrFemale == "male":
            actorChoiceList = ["A","B","C","D","E"]
            actorChoice = input("Please choose from the following celebrities to find out their respected diets by typing the "
                   "letter next to their name: \nA. Christian Bale \nB. Hugh Jackman \nC. Jonah Hill \nD. Terry Crews "
                   "\nE. Mathew McConaughey")
            actorChoice = actorChoice.upper()
            if actorChoice not in actorChoiceList:
                print("Your choice is not in the list ('A', 'B', 'C', 'D', or 'E')")
                print("Please try again")
            else:
                invalidInput = False
        elif maleOrFemale == "female":
            actorChoiceList = ["1","2","3","4","5"]
            actorChoice = input("Please choose from the following celebrities to find out their respected diets by typing the "
                   "number next to their name: \n1. Beyonce \n2. Rihanna \n3. Pippa Middleton \n4. Marilyn Monroe "
                   "\n5. Jennifer Lopez")
            if actorChoice not in actorChoiceList:
                print("Your choice is not in the list (1, 2, 3, 4, or 5)")
                print("Please try again")
            else:
                invalidInput = False

    input("Press ENTER to continue")
    print("Now that you selected your preferred celebrity, please input your age (in years), height (in inches), and weight(in lbs) to find out if this diet is ideal for you!")
    invalidInput = True
    while invalidInput:
        try:
            age = int(input("Age: "))
            invalidInput = False
        except:
            print("The input you provided is not a number.  Please try again.")

    invalidInput = True
    while invalidInput:
        try:
            height = int(input("Height: "))
            invalidInput = False
        except:
            print("The input you provided is not a number.  Please try again.")

    invalidInput = True
    while invalidInput:
        try:
            weight = int(input("Weight: "))
            invalidInput = False
        except:
            print("The input you provided is not a number.  Please try again.")

    return name, maleOrFemale, actorChoice, age, height, weight


def dietProcessing(name, maleOrFemale, actorChoice, age, height, weight):
    #print("processing")
    if maleOrFemale == "male":
        if actorChoice =="A":
            #print("Christian Bale - Batman Diet")
            dietAnswer = "Christian Bale - Batman Diet"
        if actorChoice =="B":
            #print("Hugh Jackman - Wolverine Diet")
            dietAnswer = "Hugh Jackman - Wolverine Diet"
        if actorChoice =="C":
            #print("Jonah Hill - Weight Loss Diet")
            dietAnswer = "Jonah Hill Diet"
        if actorChoice =="D":
            #print("Terry Crews - Intermittent Fasting")
            dietAnswer = "Terry Crews - Intermittent Fasting"
        if actorChoice =="E":
            #print("Mathew McConaughey - Paleo Diet")
            dietAnswer = "Mathew McConaughey - Paleo Diet"
    #print("processing")
    elif maleOrFemale == "female":
        if actorChoice == "1":
            # print("Beyonce - Herculean Diet")
            dietAnswer = "Beyonce - Herculean Diet"
        if actorChoice == "2":
            # print("Rihanna - 5-Factor Diet")
            dietAnswer = "Rihanna - 5-Factor Diet"
        if actorChoice == "3":
            # print("Pippa Middleton - Dukan Diet")
            dietAnswer = "Pippa Middleton - Dukan Diet"
        if actorChoice == "4":
            # print("Marilyn Monroe - Diuretic Diet")
            dietAnswer = "Marilyn Monroe - Diuretic Diet"
        if actorChoice == "5":
            # print("Jennifer Lopez - Intermittent Diet")
            dietAnswer = "Jennifer Lopez - Intermittent Diet"

    print("DEBUG: dietAnswer:",dietAnswer)

    #Get diet health info from file here
    #Include the above info in the healthyChoice variable

    healthFile = open("dietHealthInfo.txt","r")
    lineCount = 0
    getData = False
    #line count = 0, it makes sure that when you read the file, each variable gets in the correct data

    for line in healthFile:
        #for loop goes through each line of the file
        if line.find(dietAnswer) != -1:
        #this if statment checks if the celebrity is in the health file
            getData = True
            #print("DEBUG: getData")
            lineCount += 1
        # print(lineCount)
        # print(getData)
        if getData and lineCount > 0:
            #here the if statment checks if geData is true and line count is > 0 it gives each variable the correct info
            #print("DEBUG: lineCount:",lineCount)
            if lineCount == 2:
            #here caloric intake is being taken from the file
                caloricIntake = line.replace("\n","")
                #print("DEBUG: caloricIntake",caloricIntake)
                lineCount += 1
            elif lineCount == 3:
                recFood = line.replace("\n","")
                #print("DEBUG: recFood",recFood)
                lineCount += 1
            elif lineCount == 5:
                healthRisk = line.replace("\n","")
                #print("DEBUG: healthRisk",healthRisk)
                lineCount += 1
            elif lineCount == 4:
                dietLength = line.replace("\n", "")
                #print("DEBUG: dietLength", dietLength)
                lineCount += 1
            elif lineCount == 6:
                dietAge = line.replace("\n","")
                dietAge = int(dietAge)
                # print("DEBUG: dietAge",dietAge)
                lineCount += 1
            elif lineCount == 7:

                dietHeight = line.replace("\n","")
                dietHeight = int(dietHeight)
                #print("DEBUG: dietHeight",dietHeight)
                lineCount += 1
            elif lineCount == 8:
                dietWeight = line.replace("\n","")
                dietWeight = int(dietWeight)
                #print("DEBUG: dietWeight",dietWeight)
                getData = False
            else:
                lineCount += 1

    if age <= dietAge or height <= dietHeight or weight <= dietWeight:
        dietFeedback = "Based on your age, height, and/or weight, this diet is risky and you should consult with your doctor before trying it."
    else:
        dietFeedback = "Your age, height, and weight seem right. You should try this diet if your doctor approves!"

    return name, dietAnswer, caloricIntake, recFood, dietLength, healthRisk, dietFeedback

def dietOutput(name, dietAnswer, caloricIntake, recFood, dietLength, healthRisk, dietFeedback):
    #print("output")
    input("Well, " + name + ", the diet associated with the celebrity you chose is: " + dietAnswer + ". (press ENTER to continue)")
    input("Here is some information about the diet (press ENTER to continue):")
    input(caloricIntake + " (press ENTER to continue)")
    input(recFood + " (press ENTER to continue)")
    input(dietLength + " (press ENTER to continue)")
    input(healthRisk + " (press ENTER to continue)")
    input(dietFeedback + " (press ENTER to continue)")
    print("Thanks for playing along!  See you next time.")


main()
```
